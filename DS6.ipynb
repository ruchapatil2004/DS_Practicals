```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.set_index("Id")
```

```python
X = df.drop(['Species'],axis = 1)
y = df['Species']
```

```python
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
import sklearn.metrics
```

```python
lbs = ['Iris-versicolor','Iris-setosa','Iris-virginica']
```

```python
sklearn.metrics.confusion_matrix(y_test, y_pred, labels = lbs)
```

```python
sklearn.metrics.accuracy_score(y_test, y_pred)
```

```python
sklearn.metrics.precision_score(y_test, y_pred, labels = lbs, pos_label = 1, average='micro')
```

```python
sklearn.metrics.f1_score(y_test, y_pred, labels=lbs, pos_label = 1, average='micro')
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)
```

```python
print(acc)
```

```python
error_rate = 1 - acc
print(error_rate)
```
